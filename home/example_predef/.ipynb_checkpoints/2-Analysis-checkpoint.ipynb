```octave
% Make sure we have the MEDA Toolbox installed, so we can add it to the path in Octave:
addpath('/MEDA-Toolbox');
addpath('/MEDA-Toolbox/BigData');
pkg load statistics
```

```octave
% Transform observation-wise csv files into comple datasets that can be imported to Matlab/Octave 
% It combines the data from all the files with names "output_xxxx" at the folder "parsing_output", and outputs 'parsed_data.mat', which includes:
% x : matrix with values of the observations and variables
% obs_l : labels (timestamps) for the observations
% var_l : labels for the variables (from the naming in the configuration file used along with the FCParser)
% weight : weights for each variable (from the weights in the configuration file used along with the FCParser) 

prepare_data('parsing_output');
```

```octave
% Load the file which includes x, obs_l, var_l & weight
parsed_data = load('parsed_data.mat');

% display a summary of the data
disp('- First 5 rows and 10 columns')
disp(parsed_data.x(1:5,1:10));
disp('- Labels for the first 5 rows')
for i=1:5, disp(parsed_data.obs_l{i}), end
disp('- Labels for the first 10 columns')
for i=1:10, disp(parsed_data.var_l{i}), end
disp('- Weights for the first 5 rows')
disp(parsed_data.weight(1:5)');


% Inicialization of some parameters
prep_x = 2 % we use autoscaling as preprocessing approach (since the nature of the variables is different, packets/IPs/ports)
max_PCs = 10 % maximum number of PCs to take into account
```

```octave
% Data Analysis with PCA (Principal Component Analysis)

% 1. Selection of the number of Principal Components (PC)

var_pca(parsed_data.x,1:max_PCs,prep_x); % We look for an optimal compromise between captured variance and parsimony: 1 PC is a good choice
pcs = 1
```

```octave
% 2. MSNM plots for outliers detection and diganosis

% detection
mspc_pca(parsed_data.x,pcs,[],prep_x); % Observation #61 is a clear outlier. This often sine from 1h, there is a 1m interval with only a handlful of non-synchronized flows
```

```octave
% diagnois
omeda_pca(parsed_data.x,pcs,parsed_data.x(61,:),[],prep_x,[],parsed_data.var_l); % Observation #61 shows lower values (negative bars) for most variables
```

```octave
% Delete outlier and repeat all over again

parsed_data.x(61,:) = [];
parsed_data.obs_l(61) = [];

% 1.b Selection of the number of Principal Components (PC)

var_pca(parsed_data.x,1:max_PCs,prep_x); % We look for an optimal compromise between captured variance and parsimony: 1 PC is a good choice
pcs = 1
```

```octave
% 2.b MSNM plots for outliers detection and diganosis

% detection
mspc_pca(parsed_data.x,pcs,[],prep_x); % Observation #26 is a clear outlier
```

```octave
% diagnois
ov = omeda_pca(parsed_data.x,pcs,parsed_data.x(26,:),[],prep_x,[],parsed_data.var_l); % Observation #26 shows higer values (positive bars) for several variables

features = parsed_data.var_l(find(ov>20)) %  More UDP traffic than expected, with small packets, in ports between 1024 and 49151
```

```octave
# validate diagnosis in the UDP traffic
varN = find(ismember(parsed_data.var_l,'protocol_udp'))

figure, bar(parsed_data.x(:,varN)) % we observe a clear peack at observation #26 
```

```octave
# validate diagnosis in the small packages
varN = find(ismember(parsed_data.var_l,'in_nbytes_low'))

figure, bar(parsed_data.x(:,varN)) % we observe again a clear peack at observation #26 
```

```octave
% 3.b prepare data for downstream

% outlier 26
features
str = parsed_data.obs_l{26};
timestamps = {[str(1:4) '-' str(5:6) '-' str(7:8) ' ' str(9:10) ':' str(11:12) ':00']}

% save it to use fcdeparsing
%save('deparsing_input','features','timestamps')

f = fopen('deparsing_input','w');
fprintf(f,'features:\n%s\n',disp(features));
fprintf(f,'timestamps:\n%s\n',disp(timestamps));
fclose(f);
```
